```python
# Step 1: Import libraries
import pandas as pd

# Step 2: Create a small anonymized dataset
data = {
    'School': [
        'Bronx Academy', 'Harlem Prep', 'Queens Tech', 'Bronx Academy', 'Manhattan Scholars', 'Riverside Prep', 'Hudson Academy', 'Eastside STEM', 'Metro Leadership School', 'Central Honors'
    ],
    'Year': [2018, 2018, 2019, 2019, 2020, 2020, 2021, 2021, 2022, 2022],
    'Attendance': [94.2, 92.8, 90.5, 89.4, 93.7, 91.5, 95.1, 92.6, 94.8, 90.9],
    'Average_Score': [72, 74, 69, 70, 75, 73, 80, 77, 78, 76]
}

# Step 3: Convert to DataFrame
df = pd.DataFrame(data)

# Step 4: Preview the dataset
df.style.set_properties(subset=['School'], **{'text-align': 'left'})
```

```python
# Step 1: General information about the dataset
print("Dataset Overview:")
print(df.info())

# Step 2: Basic descriptive statistics
print("\n Summary Statistics:")
display(df.describe().round(2))

# Step 3: Check for missing or null values
print("\n Missing Values:")
print(df.isna().sum())

# Step 4: Quick unique value check (optional)
print("\n Unique Schools:")
print(df['School'].unique())

# Step 5: Correlation between attendance and performance
print("\n Correlation Matrix:")
display(df.corr(numeric_only=True).round(2))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set clean, professional style
sns.set(style="whitegrid", palette="muted")
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=df, x='Year', y='Attendance', marker='o')
plt.title("Average Attendance Trend (2018-2022)", fontsize=14, weight='bold')
plt.ylabel("Attendance (%)")
plt.xlabel("Year")
plt.savefig("Attendance_Trend.png", dpi=300, bbox_inches='tight')
plt.show()

plt.figure(figsize=(8,5))
sns.lineplot(data=df, x='Year', y='Average_Score', color='skyblue')
plt.title("Average Student Performance by Year", fontsize=14, weight='bold')
plt.ylabel("Average Score")
plt.xlabel("Year")
plt.savefig("Average_Score_Trend.png", dpi=300, bbox_inches='tight')
plt.show()

plt.figure(figsize=(6,6))
sns.regplot(data=df, x='Attendance', y='Average_Score', scatter_kws={'s':70}, line_kws={'color':'red'})
plt.title("Attendance vs. Academic Performance", fontsize=14, weight='bold')
plt.ylabel("Average Score")
plt.xlabel("Attendance (%)")
plt.savefig("Attendance_Performance_Correlation.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Combined dashboard visualization (clean version)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Chart 1: Attendance Trend
sns.lineplot(data=df, x='Year', y='Attendance', marker='o', ci=None, ax=axes[0])
axes[0].set_title("Average Attendance Trend (20182022)", fontsize=12, weight='bold')
axes[0].set_ylabel("Attendance (%)")
axes[0].set_xlabel("Year")

# Chart 2: Student Performance Trend
sns.lineplot(data=df, x='Year', y='Average_Score', marker='o', ci=None, ax=axes[1], color='orange')
axes[1].set_title("Average Student Performance by Year", fontsize=12, weight='bold')
axes[1].set_ylabel("Average Score")
axes[1].set_xlabel("Year")

# Chart 3: Attendance vs. Performance
sns.regplot(data=df, x='Attendance', y='Average_Score', ax=axes[2], color='green')
axes[2].set_title("Attendance vs. Academic Performance", fontsize=12, weight='bold')
axes[2].set_ylabel("Average Score")
axes[2].set_xlabel("Attendance (%)")

plt.tight_layout()
plt.savefig("Python_StudentAchievement_Dashboard.png", dpi=300, bbox_inches='tight')
plt.show()
```
