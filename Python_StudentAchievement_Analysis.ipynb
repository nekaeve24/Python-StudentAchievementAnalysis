```python
# =============================================
# Python Student Achievement Analysis
# Complete Script (Compact Dashboard Version)
# =============================================

# --- Import required libraries ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter

# =============================================
# STEP 1: Create a Synthetic Dataset (Safe & Anonymized)
# =============================================
data = {
    'School': [
        'North Academy', 'East STEM High', 'West Prep', 'South Tech',
        'Central Honors', 'Hillside Collegiate', 'Riverside Institute',
        'Metro Leaders', 'Summit Scholars', 'Global Prep'
    ],
    'Year': [2018, 2019, 2020, 2021, 2022, 2018, 2019, 2020, 2021, 2022],
    'Attendance': [93.4, 90.0, 92.5, 94.9, 93.0, 91.7, 89.4, 92.1, 95.1, 90.9],
    'Average_Score': [72.3, 69.1, 74.5, 78.3, 76.9, 71.8, 70.4, 73.0, 80.0, 75.7]
}

df = pd.DataFrame(data)

# =============================================
# STEP 2: Explore & Describe Data
# =============================================
print("üßæ Dataset Overview:")
print(df.info())

print("\nüìä Summary Statistics:")
print(df.describe().round(2))

print("\nüîç Missing Values:")
print(df.isna().sum())

print("\nüè´ Unique Schools:")
print(df['School'].unique())

print("\nüìà Correlation Matrix:")
print(df.corr(numeric_only=True).round(2))

# =============================================
# STEP 3: Data Cleaning & Visualization Setup
# =============================================
sns.set(style="whitegrid", palette="muted")

# Clean numeric formatting for visuals
df['Attendance'] = df['Attendance'].round(1)       # One decimal
df['Average_Score'] = df['Average_Score'].round(1) # One decimal
df['Year'] = df['Year'].astype(int)                # Whole numbers

plt.close('all')

# =============================================
# STEP 4: Combined Dashboard (Three Charts Side by Side)
# =============================================
fig, axes = plt.subplots(1, 3, figsize=(9,3))  # smaller layout

# Chart 1: Attendance Trend
sns.lineplot(data=df, x='Year', y='Attendance', marker='o', color='steelblue', ax=axes[0])
axes[0].set_title("Attendance Trend", fontsize=10, weight='bold')
axes[0].set_xlabel("Year")
axes[0].set_ylabel("Attendance (%)")
axes[0].xaxis.set_major_formatter(FuncFormatter(lambda x, _: f"{int(x)}"))
axes[0].yaxis.set_major_formatter(FuncFormatter(lambda y, _: f"{y:.1f}"))

# Chart 2: Student Performance by Year
sns.lineplot(data=df, x='Year', y='Average_Score', color='orange', marker='o', ax=axes[1])
axes[1].set_title("Performance by Year", fontsize=10, weight='bold')
axes[1].set_xlabel("Year")
axes[1].set_ylabel("Average Score")
axes[1].xaxis.set_major_formatter(FuncFormatter(lambda x, _: f"{int(x)}"))
axes[1].yaxis.set_major_formatter(FuncFormatter(lambda y, _: f"{y:.1f}"))

# Chart 3: Attendance vs. Performance
sns.regplot(data=df, x='Attendance', y='Average_Score', color='crimson', line_kws={"color":"black"}, ax=axes[2])
axes[2].set_title("Attendance vs. Performance", fontsize=10, weight='bold')
axes[2].set_xlabel("Attendance (%)")
axes[2].set_ylabel("Average Score")
axes[2].xaxis.set_major_formatter(FuncFormatter(lambda x, _: f"{x:.1f}"))
axes[2].yaxis.set_major_formatter(FuncFormatter(lambda y, _: f"{y:.1f}"))

plt.tight_layout()
plt.savefig("Python_StudentAchievement_Dashboard.png", dpi=300, bbox_inches='tight')
plt.show()
```
